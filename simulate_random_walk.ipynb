# Simulate a random walk for 2-dimensional diffusion

```python
import numpy as np
import matplotlib.pyplot as plt
import random

def randomwalk2D(n):
    # [0, 0, 0, ... ,0]
    x = np.zeros(n)
    y = np.zeros(n)

    directions = ["UP", "DOWN", "LEFT", "RIGHT"]
    for i in range(1, n):
        # Pick a direction at random
        step = random.choice(directions)
        
        # Move the object according to the direction
        if step == "RIGHT":
            x[i] = x[i - 1] + 1
            y[i] = y[i - 1]
        elif step == "LEFT":
            x[i] = x[i - 1] - 1
            y[i] = y[i - 1]
        elif step == "UP":
            x[i] = x[i - 1]
            y[i] = y[i - 1] + 1
        elif step == "DOWN":
            x[i] = x[i - 1]
            y[i] = y[i - 1] - 1
    
    # Return all the x and y positions of the object
    return x, y

x_data, y_data = randomwalk2D(2000)

plt.title("2D Random Walk in Python")
plt.plot(x_data, y_data)
plt.show()
```

```python
def  randomWalk2D_CoinFlip(n):
    x=np.zeros(n)
    y=np.zeros(n)
    for i in range(1,n):
        p_x = np.random.rand(1)
        p_y = np.random.rand(1)
        if p_x<0.33:
            x[i]=-1
        if 0.66<p_x>=0.33:
            x[i]=0
        if p_x>=0.66:
            x[i]=1
        if p_y<0.33:
            y[i]=-1
        if 0.66<p_y>=0.33:
            y[i]=0
        if p_y>=0.66:
            y[i]=1
        #Update the step
        x[i]=x[i-1]+x[i]
        y[i]=y[i-1]+y[i]
    return x,y
```

```python
step_x,step_y=randomWalk2D_CoinFlip(10000)
```

```python
plt.title("2D Random Walk in Python")
plt.plot(step_x, step_y)
plt.show()
```
